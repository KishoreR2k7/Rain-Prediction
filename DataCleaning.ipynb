```python
import numpy as np
import pandas as pd
df=pd.read_csv('usa_rain_prediction_dataset_2024_2025.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
df.duplicated().sum()
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df['Month']=df['Date'].dt.month
df['Day']=df['Date'].dt.day
df['Year']=df['Date'].dt.year
df.drop('Date',axis=1,inplace=True)
df.head()
```

```python
df['Location'].value_counts().index
```

```python
df=pd.get_dummies(df,columns=['Location'],drop_first=True,dtype=int)
```

```python
df.head()
```

```python
df.dropna(subset=['Cloud Cover'], inplace=True)
df.head(15)
```

```python
df.info()
```

```python
df['Cloud Cover']=pd.to_numeric(df['Cloud Cover'], errors='coerce')
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr=correlation(df,0.85)
corr
```

```python
df.to_csv('Cleaned_Data.csv',index=False)
```
