```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Cleaned_Data.csv')
df.head()
```

```python
monthly_rain=df.groupby('Month')['Rain Tomorrow'].sum().reset_index()
sns.barplot(x='Month',y='Rain Tomorrow',data=monthly_rain)
```

```python
monthly_rain=df.groupby('Month')['Rain Tomorrow'].sum()
monthly_rain.plot(kind='line',marker='o')
plt.title('Month wise Rain')
```

```python
import matplotlib.pyplot as plt
counts = df['Rain Tomorrow'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(counts,labels=['No Rain', 'Rain'],autopct='%1.1f%%', colors=['skyblue','orange'])
plt.title('Rain Tomorrow Distribution')
plt.show()
```

```python
sns.countplot(x='Rain Tomorrow', data=df)
plt.title('Rain Tomorrow Distribution')
plt.xlabel('Rain Tomorrow')
plt.ylabel('Count')
plt.show()
```
